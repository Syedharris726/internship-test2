```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import csv 

#Loading dataset
df = pd.read_csv('/ma.csv')
for i in range(0,54,10):
  for j in range(1,10):
    if i+j>54:
      break
    else:
      df.loc[i,'Population']+=df.loc[i+j,'Population']
      df.loc[i,'Total']+=df.loc[i+j,'Total']
      df.loc[i,'Violent']+=df.loc[i+j,'Violent']
      df.loc[i,'Property']+=df.loc[i+j,'Property']
      df.loc[i,'Murder']+=df.loc[i+j,'Murder']
      df.loc[i,'Forcible_Rape']+=df.loc[i+j,'Forcible_Rape']
      df.loc[i,'Robbery']+=df.loc[i+j,'Robbery']
      df.loc[i,'Aggravated_assault']+=df.loc[i+j,'Aggravated_assault']
      df.loc[i,'Burglary']+=df.loc[i+j,'Burglary']
      df.loc[i,'Larceny_Theft']+=df.loc[i+j,'Larceny_Theft']
      df.loc[i,'Vehicle_Theft']+=df.loc[i+j,'Vehicle_Theft']    

  
# writing into the file
df.to_csv("/ma.csv", index=False)
  
print(df)

```
